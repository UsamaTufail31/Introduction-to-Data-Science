```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
df=pd.read_csv("gender-prediction.csv");
```

```python
df['gender'].value_counts()
```

```python
labels = preprocessing.LabelEncoder()
df['beard'] = labels.fit_transform(df['beard'])
df['hair_length'] = labels.fit_transform(df['hair_length'])
df['eye_color'] = labels.fit_transform(df['eye_color'])
df['scarf'] = labels.fit_transform(df['scarf'])
df['gender'] = labels.fit_transform(df['gender'])
```

```python
x = df.drop('gender', axis=1)
y = df['gender']
```

**Logistic Regression**

```python
#Logistic Regression
X_train, x_test, Y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 2,)
model = LogisticRegression()
model.fit(X_train, Y_train)
y_pred = model.predict(x_test)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(classification_rep)
```

**Monte Carlo cross-validation** and **Leave P-Out cross-validation**

```python
def monte_carlo_cross_validation(model, X, y, iterations=100):
    f1_scores = []
    for _ in range(iterations):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, )
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        f1_scores.append(f1_score(y_test, y_pred, ))
    return np.mean(f1_scores)
def leave_p_out_cross_validation(model, X, y, p=5):
    fold = StratifiedKFold(n_splits=p)
    f1_scores = []
    for train, test in fold.split(X, y):
        X_train, X_test = X.iloc[train], X.iloc[test]
        y_train, y_test = y.iloc[train], y.iloc[test]
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        f1_scores.append(f1_score(y_test, y_pred,))
    return np.mean(f1_scores)

rf_classifier = RandomForestClassifier(random_state=42)
monte_carlo_f1 = monte_carlo_cross_validation(rf_classifier, X, y)
leave_p_out_f1 = leave_p_out_cross_validation(rf_classifier, X, y, p=4)
print(f'Monte Carlo Cross-Validation F1 Score: {monte_carlo_f1:.4f}')
print(f'Leave P-Out Cross-Validation F1 Score: {leave_p_out_f1:.4f}')
```
