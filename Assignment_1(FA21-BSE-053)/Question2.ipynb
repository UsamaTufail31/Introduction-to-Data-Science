```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url='https://www.timeanddate.com/on-this-day/december/2'
doc = requests.get(url, headers={'Accept-Language': 'en-gb',	'User-Agent': ''})
```

```python
soup=bs(doc.content,'html.parser')
```

```python
smallSoup= soup.find('div',{'class':'otd-life__block'}).find_all('h3',{'class':'otd-title'})
```

```python
names=[]
for person in smallSoup:
  names.append(person.text)
```

```python
for x in range(len(names)):
  names[x]=names[x][5:]
```

```python
with open("output.txt","w") as f:
  f.write("Name of Persons who share their birth day:\n\n")
  for y in names:
    f.write(y+" \n")
```

```python
url2='https://www.britannica.com/on-this-day/December-2'
doc2=requests.get(url2,headers={'Accept-Language': 'en-gb',	'User-Agent': ''})
```

```python
soup2=bs(doc2.content,'html.parser')
```

```python
events=[]
```

```python
events.append('Event 1 :'+soup2.find('div',{'class':'description font-serif'}).text)
```

```python
allportions=soup2.find_all('div',{'class':'card-body font-serif'})
```

```python
count=2
for one in allportions:
  events.append('Event '+ str(count)+' : '+one.text.strip())
  count=count+1
```

```python
with open('output.txt', 'a') as file:
  file.write("\n\nEvents on my Birthday(2-Dec):\n\n")
  for item in events:
    file.write(item+"\n")
```
